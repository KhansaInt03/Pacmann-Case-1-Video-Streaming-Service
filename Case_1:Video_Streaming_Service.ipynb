```python
from tabulate import tabulate
```

```python
# User Data
data = {
    "Shandy": ["Basic Plan", 12, "shandy-2134"],
    "Cahya": ["Standard Plan", 24, "cahya-abcd"],
    "Ana": ["Premium Plan", 5, "ana-2f9g"],
    "Bagus": ["Basic Plan", 11, "bagus-9f92"]
}
```

```python
# Create User Class
class User:

  data = {
    "Shandy": ["Basic Plan", 12, "shandy-2134"],
    "Cahya": ["Standard Plan", 24, "cahya-abcd"],
    "Ana": ["Premium Plan", 5, "ana-2f9g"],
    "Bagus": ["Basic Plan", 11, "bagus-9f92"]
}

  plan_level = { "Basic Plan": 0, "Standard Plan": 1, "Premium Plan": 2 }

  def __init__(self, username, duration:int, plan):
    self.username = username
    self.duration_plan = duration
    self.current_plan = plan

  def check_benefit(self):
    self.pacflix_plan = {
        "Basic Plan": [True, True, True, False, False, 1, "3rd party movie only", 120_000],
        "Standard Plan": [True, True, True, True, False, 2, 'Basic Plan Content + Sports', 160_000],
        "Premiun Plan": [True, True, True, True, True, 4, "Basic Plan + Standard Plan + Pacflix Original Series", 200_000],
        "Services": ["Stream", "Download", "SD Quality", "HD Quality", "UHD Quality", "Number of Device", "Content", "Price (IDR)"],
    }
    table_plan = tabulate(self.pacflix_plan, headers = "keys", tablefmt = "grid", stralign = "center")
    print("PacFlix Plan List\n")
    print(table_plan)

  def check_plan(self, username):

    self.username = username

    if self.username in self.data.keys():
      user_data = self.data.get(self.username) # getting data with key
      user_plan, user_duration = user_data[0], user_data[1]
      get_pacflix_plan = { user_plan: self.pacflix_plan[user_plan], "Services": self.pacflix_plan["Services"]}
      print(f"""
      User {self.username}'s Data

      Plan: {user_plan}
      Duration: {user_duration} month

      {user_plan} PacFlix Benefit List
            """)
      print(tabulate(get_pacflix_plan, headers = "keys", tablefmt = "grid", stralign = "center"))

    else:
      print("User not found")

  def ugrade_plan(self, username, current_plan, new_plan):
    self.username = username
    self.current_plan = current_plan
    self.new_plan = new_plan
    self.fin_price = 0

    if self.username in self.data.keys():
      duration = self.data.get(self.username)[1] # getting user's duration
      if self.plan_level[self.current_plan] < self.plan_level[self.new_plan]:
        if duration > 12:
          self.fin_price = self.pacflix_plan[self.new_plan][-1] - ( 5/100 * self.pacflix_plan[self.new_plan][-1] )
        else:
          self.fin_price = self.pacflix_plan[self.new_plan][-1]
      else:
        print("You cannot downgrade PacFlix Plan.")
    else:
      print("Username is not found.")

    print(f"Price you need to pay: {self.fin_price}")
```

```python
# Create Object for User Class
user_1 = User("Shandy", 12, "Basic Plan")
```

```python
# Check attribute in Class
user_1.username, user_1.duration_plan, user_1.current_plan
```

```python
# Test Case 1 -> Showing plan with the benefit + price
user_1.check_benefit()
```

```python
# Test case 2 -> Showing user's plan benefit and duration using username only

user_1.check_plan("Shandy")
```

```python
# Test case 3 -> Upgrading plan, if duration > 12, discount 5%. Can't downgrade

user_1.ugrade_plan("Ana", "Premium Plan", "Standard Plan")
```
